```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.sharesansar.com/today-share-price','html.parser')
raw_code = requests.get(url)
raw_code.text
```

```python
lxml_code = BeautifulSoup(raw_code.text, "lxml")
lxml_code
```

```python
table_code = lxml_code.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
data = []
for tag in tags:
    y = tag.text.split('\n')
    if y[1] != '':
        data.append(y[1:-1])
    
print(data)
```

```python
import csv
with open('Today Share Price.csv','w') as file:
    x = csv.writer(file)
    x.writerows(data)
```
